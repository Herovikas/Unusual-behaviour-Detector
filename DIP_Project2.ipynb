```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
Dir = "Images"
X_usual = []
y_usual = []
labels = ["Images_usual"]
def creating_usual_training_data():
    for category in labels:
        path = os.path.join(Dir, category)
        class_number = labels.index(category)
        for imvar in os.listdir(path):
            image = cv2.imread(os.path.join(path,imvar), cv2.IMREAD_GRAYSCALE)
            try:
                image = cv2.resize(image, (300, 300))
            except:
                pass
            image_array = np.array(image).flatten()
            X_usual.append(image_array)
            y_usual.append(class_number)

    
creating_usual_training_data()                                         

#type(X)
```

```python
X_unusual = []
y_unusual = []
labels = ["Images_unusual"]
def creating_unusual_training_data():
    for category in labels:
        path = os.path.join(Dir, category)
        class_number = labels.index(category)
        for imvar in os.listdir(path):
            image = cv2.imread(os.path.join(path,imvar), cv2.IMREAD_GRAYSCALE)
            try:
                image = cv2.resize(image, (300, 300))
            except:
                pass
            image_array = np.array(image).flatten()
            X_unusual.append(image_array)
            y_unusual.append(class_number + 1)

    
creating_unusual_training_data()                                         

#type(X)
```
